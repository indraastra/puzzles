```python
import math
```

```python
pandigital_sum = sum(range(1, 10))

def is_pandigital(n):
    digits = set(str(n))
    return '0' not in digits and len(digits) == 9

def num_digits(n):
    return math.floor(math.log10(n)) + 1

def has_pandigital_start(n):
    return is_pandigital(n // (10 ** max(0, num_digits(n)-9)))

def has_pandigital_end(n):
    return is_pandigital(n % 1_000_000_000)
```

```python
%%time
fibs = (1, 1)

for i in range(3, 1_000_000):
    f = sum(fibs)
    if has_pandigital_end(f) and has_pandigital_start(f):
        print(i)
        break
    fibs = (fibs[-1], f)
```

```python
import math
```

```python
## Slow version:
blue = 2
prev_blue = 1
found = 0
limit = 1_000_000

while True:
    num = 2*blue*(blue-1)
    total = math.ceil(math.sqrt(num))
    if total * (total - 1) == num:
        red = total - blue
        print(f'> blue: {blue}, red: {red}, total: {total}')
        print()
        print(f'{blue} * {blue-1}'.rjust(20), '1'.rjust(3))
        print('-'*20, '=', '-')
        print(f'{total} * {total-1}'.rjust(20), '2'.rjust(3))
        print()
        print(blue/prev_blue, blue/red, blue/total, '\n')
        found += 1
        prev_blue = blue
        #blue = int(blue*(2.999+math.sqrt(8)))
        blue += 1
        if total >= limit:
            break
    blue += 1
```

```python
## Fast version:
blue = 85
prev_blue = 15
found = 0
limit = 1_000_000_000_000

while True:
    num = 2*blue*(blue-1)
    total = math.ceil(math.sqrt(num))
    if total * (total - 1) == num:
        red = total - blue
        print(f'> blue: {blue}, red: {red}, total: {total}')
        print()
        print(f'{blue} * {blue-1}'.rjust(30), '1'.rjust(3))
        print('-'*30, '=', '-')
        print(f'{total} * {total-1}'.rjust(30), '2'.rjust(3))
        print()
        print(blue/prev_blue, blue/red, blue/total, '\n')
        found += 1
        prev_blue = blue
        blue = math.ceil(blue*(3+math.sqrt(8)))
        if total >= limit:
            break
    else:
        blue -= 1
```
